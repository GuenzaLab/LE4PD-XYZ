```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
corr_list = []
for a in range(1, 11):
    corr = np.loadtxt('../mode_corr_' + str(a))
    corr_list.append(integrate.trapz(corr[:,1], x = corr[:,0]))
    print(corr_list[a - 1])
    plt.plot(corr[:, 0], corr[:, 1], lw = 2, c ='k')
    plt.axhline(0, lw = 1, ls = ':', color = 'grey')
    plt.xscale('log')
    plt.xlabel('t, ps', fontsize = 16)
    plt.ylabel('acf', fontsize = 16)
    plt.show()
    plt.close()
```

```python
msm_list = [6515, 4551, 4345, 958, 5531, 3050, 7419, 41.7, 1321, 400]
lag_list = [2800, 1500, 2000, 300, 4900, 2000, 3000, 10, 500, 200]
tau_list = np.loadtxt('../tau_scaled_3N.dat')[:, 1][:10]
```

```python
plt.plot(corr_list, lw = 0, marker = 'o', c = 'k')
plt.plot(msm_list, lw = 0, marker = 'o', c = 'r')
plt.plot(tau_list, lw = 0, marker = 'o', c = 'b')
plt.yscale('log')
plt.show()
plt.close()
```
