```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def gauss(x, var):
    return (1 / np.sqrt(2*np.pi*var))*np.exp(-x**2 / (2*var))
```

```python
covar = np.loadtxt('../com_covar_eig')
```

```python
(1 / np.sqrt(2*np.pi*covar))*np.exp(-x**2 / (2*covar))
```

```python
for i in range(25):
    pc = np.load("../xi_" + str(i + 1) + ".npy")

    histo, bins = np.histogram(pc, bins = 100)

    dx = bins[1] - bins[0]
    histo = histo / (histo.sum() * dx)

    bins = (bins[1:] + bins[:-1]) / 2
    
    plt.plot(bins, histo, lw = 2, color = 'k')
    plt.plot(bins, gauss(bins, covar[i]), color = 'r', lw = 2)
    plt.show()
    plt.close()
```
