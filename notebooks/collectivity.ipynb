```python
import numpy as np
import matplotlib.pyplot as plt
import os
print(os.getcwd())
N=76
```

```python
Q=np.reshape(np.loadtxt('../com_Rmatrix'),(N*3,N*3))
kappa=np.zeros(len(Q[:,0]))
```

```python
#Just double checking if the q_i are normalized
for i in range(len(Q[:,0])):
    print(np.sum(Q[:,i]**2))
```

```python
for i in range(3*N):
    summa=0.0
    for n in range(0, 3*N, 3):
        summa += ((Q[n,i]**2 + Q[n + 1,i]**2 + Q[n + 2,i]**2)*np.log(Q[n,i]**2 + Q[n + 1,i]**2 + Q[n + 2,i]**2))
    kappa[i]= (1/N)*np.exp(-summa)
```

```python
colors = plt.cm.gnuplot(np.linspace(kappa.min(),kappa.max(),N-4))

plt.plot(np.arange(1, 223),kappa[:-6],lw=1,color='k',ls=':',zorder=-10)
plt.scatter(np.arange(1, 223),kappa[:-6],c=kappa[:-6],cmap='cool')
plt.xlabel('Mode Number',fontsize=16)
plt.ylabel(r'$\kappa$',fontsize=16)
#plt.xticks([4,10,20,30,40,50,60,70],size=12)
plt.yticks(size=12)
#plt.xscale('log')
plt.savefig('../figures/eigenvector_collectivity.pdf',dpi=300)
plt.savefig('../figures/eigenvector_collectivity.png',dpi=300)
plt.show()
```
