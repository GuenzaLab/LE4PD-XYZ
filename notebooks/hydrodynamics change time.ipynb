```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
avresrad_list = []

for i, k in enumerate([1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]):
    if k == 1000:
        avresrad_list.append(np.loadtxt('../avresrad').mean())
    else:
        avresrad_list.append(np.loadtxt('../' + str(k) + 'ns/avresrad').mean())
```

```python
times = [1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

plt.plot(times, avresrad_list, lw = 0, marker = 'o', c = 'k')
plt.xlabel('simulation time, ns', fontsize = 16)
plt.ylabel(r'$\overline{r}_w,\ \AA$', fontsize = 16)
plt.xticks(times, size = 12)
plt.yticks([2.0, 2.04, 2.08, 2.12, 2.16, 2.2], size = 12)
plt.xscale('log')
plt.ylim((2, 2.2))
plt.savefig('figures/avresrad_change_time.pdf', dpi = 300)
plt.show()
plt.close()
```

```python
avfr_list = []

for i, k in enumerate([1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]):
    if k == 1000:
        avfr_list.append(np.loadtxt('../avfr')*1e11)
    else:
        avfr_list.append(np.loadtxt('../' + str(k) + 'ns/avfr')*1e11)
```

```python
times = [1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

plt.plot(times, avfr_list, lw = 0, marker = 'o', c = 'k')
plt.xlabel('simulation time, ns', fontsize = 16)
plt.ylabel(r'$\overline{\zeta} \times 10^{11}\ \frac{kg}{s}$', fontsize = 16)
plt.xticks(times, size = 12)
#plt.yticks([2.0, 2.04, 2.08, 2.12, 2.16, 2.2], size = 12)
plt.xscale('log')
plt.ylim((1.45, 1.55))
plt.savefig('figures/avfr_change_time.pdf', dpi = 300)
plt.show()
plt.close()
```

```python
fric_list = []

for i, k in enumerate([1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]):
    if k == 1000:
        fric_list.append(np.loadtxt('../fric')*1e2)
    else:
        fric_list.append(np.loadtxt('../' + str(k) + 'ns/fric')*1e2)
```

```python
times = [1, 10, 25, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

colors = plt.cm.cool(np.linspace(0, 1, len(times)))
for i, k in enumerate(times):
    plt.plot(fric_list[i][1:,1], lw = 2 , c = colors[i], label = str(k) + 'ns')
plt.xlabel('sequence', fontsize = 16)
plt.ylabel(r'$\zeta_i,\ \ \frac{kg}{s}$', fontsize = 16)
plt.xticks(ticks = np.arange(0, 76, 10), labels = np.arange(1, 77, 10), size = 12)
#plt.yticks([2.0, 2.04, 2.08, 2.12, 2.16, 2.2], size = 12)
#plt.xscale('log')
#plt.ylim((2, 2.2))
#plt.legend()
plt.savefig('figures/gamma_change_time.pdf', dpi = 300)
plt.show()
plt.close()
```

```python
times = [1, 10, 100, 400, 800, 1000]

colors = plt.cm.cool(np.linspace(0, 1, len(times)))
fig, axs = plt.subplots(3, 2, figsize = (12, 12), sharex = True)
counter = 0
for i, k in enumerate(times):
    print(counter)
    if i <= 2:
        if k == 1000:
            axs[counter, 0].imshow(np.reshape(np.loadtxt('../Rij'), (76, 76)))
            axs[counter, 0].set_title('1000 ns')
            axs[counter, 0].set_xticks([])
            axs[counter, 0].set_yticks([])
        else:
            axs[counter, 0].imshow(np.reshape(np.loadtxt('../' + str(k) + 'ns/Rij'), (76, 76)))
            axs[counter, 0].set_title(str(k) + ' ns')
            axs[counter, 0].set_xticks([])
            axs[counter, 0].set_yticks([])
    else:
        if k == 1000:
            axs[counter, 1].matshow(np.reshape(np.loadtxt('../Rij'), (76, 76)))
            axs[counter, 1].set_title('1000 ns')
            axs[counter, 1].set_xticks([])
            axs[counter, 1].set_yticks([])
        else:
            axs[counter, 1].matshow(np.reshape(np.loadtxt('../' + str(k) + 'ns/Rij'), (76, 76)))
            axs[counter, 1].set_title(str(k) + ' ns')
            axs[counter, 1].set_xticks([])
            axs[counter, 1].set_yticks([])
            
    if i == 2:
        counter = 0
    else:
        counter += 1
        
plt.tight_layout()
plt.savefig('figures/Rij_inv_change_time.pdf', dpi = 300)
plt.show()
plt.close()
```
