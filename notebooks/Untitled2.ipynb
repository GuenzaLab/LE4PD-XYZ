```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = '../'
PATH_LE4PD = '../../lml/'

N = 76

lml = np.zeros((N, 3*N))
lml_HA = np.zeros((N, 3*N))
lml_le4pd = np.zeros((N, 3*N))

for i in range(1,3*N-6):
    lml[:,i - 1] = np.loadtxt(PATH + 'lml_' + str(i) + '.dat')
    lml_HA[:, i - 1] = np.loadtxt(PATH + 'HC_lml_' + str(i) + '.dat')
    #dummy = np.loadtxt(PATH_LE4PD + 'lml_' + str(i + 3) + '.dat')
    #lml_le4pd[1:, i - 1] = np.loadtxt(PATH_LE4PD + 'lml_' + str(i + 3) + '.dat')
    #lml[0, i - 1] = dummy[0]
```

```python
tot_lml = lml.sum(1)
```

```python
tot_lml_HA = lml_HA.sum(1)
```

```python
frac_lml = np.zeros((N, 3*N - 6))
frac_lml_HA = np.zeros((N, 3*N - 6))
for i in range(3*N - 6):
    frac_lml[:, i] = lml[:, i] / tot_lml
    frac_lml_HA[:, i] = lml_HA[:,i] / tot_lml_HA
```

```python
for i in range(10):
    plt.plot(np.arange(1, 77), frac_lml[:,i], lw = 2, c = 'k')
    plt.plot(np.arange(1, 77), frac_lml_HA[:,i], lw = 2, c = 'b')
    plt.title('Mode ' + str(i+1))
    plt.show()
    plt.close()
```
