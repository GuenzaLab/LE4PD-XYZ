```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 76
fric = 1.526e-11 #kg / s
kT = 1.38e-23*300 #kg m^2 / s^2
b = 0.365e-9 #m
eta = fric / (6*np.pi*b)

p = np.linspace(1, 76, 100)
```

```python
tr = (fric*N**2*b**2 / (3*np.pi**2*kT*p**2))*1e9 #Rouse times in ns
```

```python
tz = (eta*N**1.5*b**3 / (np.sqrt(3*np.pi)*kT*p**1.5))*1e9 #Zimm times in ns 
```

```python
plt.plot(p, tr, lw = 2, c = 'r')
plt.plot(p, tz, lw = 2, c = 'b')
plt.yscale('log')
plt.xscale('log')
plt.show()
plt.close()
```
