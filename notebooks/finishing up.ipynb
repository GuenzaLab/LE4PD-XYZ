```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fmad = np.loadtxt('../fmad_mp_60_3N.dat')
fmad_HA = np.loadtxt('../fmad_mp_60_3N_HA.dat')
```

```python
plt.plot(np.arange(1, len(fmad)+1), fmad, lw = 0.5, color = 'k', marker = 'o', ls = ':')
plt.plot(np.arange(1, len(fmad_HA)+1), fmad_HA, lw = 0.5, color = 'r', marker = 'o', ls = ':')
plt.xscale('log')
plt.show()
plt.close()
```

```python
bar = np.zeros(10)
bar_HA = np.zeros(10)
for i in range(1, 11):
    bar[i - 1] = float(np.loadtxt('../barrier_bottom' + str(i)))
    bar_HA[i - 1] = float(np.loadtxt('../HA_barrier_bottom' + str(i)))
```

```python
plt.plot(np.arange(1, len(fmad)+1), fmad, lw = 0.5, color = 'k', marker = 'o', ls = ':')
plt.plot(np.arange(1, len(fmad_HA)+1), fmad_HA, lw = 0.5, color = 'r', marker = 'o', ls = ':')
plt.plot(np.arange(1, len(bar)+1), bar, lw = 0.5, color = 'b', marker = 'o', ls = ':')
plt.plot(np.arange(1, len(bar_HA)+1), bar_HA, lw = 0.5, color = 'magenta', marker = 'o', ls = ':')
plt.xscale('log')
plt.xlim((0.9, 11))
plt.show()
plt.close()
```
